```python
import numpy as np
import pandas as pd
```

```python
sim_rem = pd.read_excel("/content/drive/MyDrive/Investigacion/paper AI/results.xlsx", sheet_name="sim_rem")
sim_rem
```

```python
sim_rem.Percentile_25.iloc[0]
```

```python
n_obs = 42  # Number of observations

# Create an initial sample
np.random.seed(0)  # For reproducibility
observations = np.random.normal(loc=sim_rem.Median.iloc[6], scale=sim_rem.StandDev.iloc[6], size=n_obs)

# Convert to a DataFrame for better handling
df = pd.DataFrame(observations, columns=["Value"])

# Adjust to approximate quartiles
# Using quantile function to get the current Q1 and Q3 from the sample
current_Q1 = df['Value'].quantile(0.25)
current_Q3 = df['Value'].quantile(0.75)

# Scale the data to fit the desired Q1, median, Q3
df['Scaled'] = ((df['Value'] - current_Q1) / (current_Q3 - current_Q1)) * (sim_rem.Percentile_75.iloc[6] - sim_rem.Percentile_25.iloc[6]) + sim_rem.Percentile_25.iloc[6]

# Adjust median
median_adjustment = sim_rem.Median.iloc[6] - df['Scaled'].median()
df['Final'] = df['Scaled'] + median_adjustment

# Take the final values
feb_25 = df['Final'].to_numpy()
feb_25
```
