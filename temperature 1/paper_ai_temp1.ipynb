````python
from openai import OpenAI

# Initialize the OpenAI client
client = OpenAI()

def get_completion(prompt, model="gpt-4o-mini", temperature=1):
    # Creating the completion using the new client object
    completion = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ],
        temperature=temperature,
    )

    # Return the response content
    return completion.choices[0].message.content
````

Sample 1: jan 2023 - jul 2023

````python
# Prompt
prompt = f"""
    Assume you are in Argentina on January 31, 2023. Also, you know that the monthly inflation for the last 6 months was: July 2022: 7.4%, August 2022: 7%, September 2022: 6.2%, October 2022: 6.3%, November 2022: 4.9%, December 2022: 5.1%. Please make your best forecast of monthly inflation based on the Consumer Price Index (CPI) – General Level Argentina, for the period January 2023 to July 2023.
    To give your answer, do not use information available in your training data after January 2023.
    Give your answer in numbers, monthly percentages. Provide them in python dictionary format with the key being the month and the value the corresponding percentage. Limit your answer to the dictionary and do not include additional information.
"""
n_times = 40  # Number of times to repeat the prompt

sample_1 = []

for i in range(n_times):
    # Get the raw response from the model
    raw_response = get_completion(prompt)

    # Clean up the response by removing triple backticks
    clean_response = raw_response.strip('```python').strip('```').strip()

    # Convert the clean JSON string into a Python dictionary and append to the list
    sample_1.append(json.loads(clean_response))

# Print the list of dictionaries
#print(sample_1)
````

````python
# Prompt
prompt = f"""
    Assume you are in Argentina on July 31, 2024. Also, you know that the monthly inflation for the last 6 months was: January 2024: 20.6%, February 2024: 13.2%, March 2024: 11%, April 2024: 8.8%, May 2024: 4.2%, June 2024: 4.6%. Please make your best forecast of monthly inflation based on the Consumer Price Index (CPI) – General Level Argentina, for the period July 2024 to January 2025.
    To give your answer, do not use information available in your training data after July 2024.
    Give your answer in numbers, monthly percentages. Provide them in python dictionary format with the key being the month and the value the corresponding percentage. Limit your answer to the dictionary and do not include additional information.
"""
n_times = 42

sample_19 = []

for i in range(n_times):
    # Get the raw response from the model
    raw_response = get_completion(prompt)

    # Clean up the response by removing triple backticks
    clean_response = raw_response.strip('```python').strip('```').strip()

    # Convert the clean JSON string into a Python dictionary and append to the list
    sample_19.append(json.loads(clean_response))
````
