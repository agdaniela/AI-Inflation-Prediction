```python

forecast_df_mean = pd.read_excel("/content/drive/MyDrive/Investigacion/paper AI/results.xlsx", sheet_name="forecasts_mean")
#forecast_df_mean.index = forecast_df_mean["dates"]
forecast_df_mean = forecast_df_mean.drop('dates', axis=1)
#forecast_df_mean.columns = pd.date_range(start= "2023-01-01"  ,periods = 26, freq ="M")

forecast_df_mean.index = pd.date_range(start= "2023-01-01"  ,periods = 20, freq ="MS")
forecast_df_mean.columns = pd.date_range(start= "2023-01-01"  ,periods = 26, freq ="MS")
```

```python
actual_inflation = df["inflation"]
forecast_df = forecast_df_mean
```

```python
import matplotlib.dates as mdates

# Plot the actual inflation data
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(actual_inflation.index, actual_inflation.values, color='rebeccapurple', linewidth=2, marker='o', label='Actual Inflation')

# Plot forecast data for each forecast period
for idx, row in forecast_df.iterrows():
    forecast_values = row.dropna()  # Exclude NaNs
    forecast_dates = pd.date_range(start=idx, periods=len(forecast_values), freq='MS')

    # Plot the forecast line
    ax.plot(forecast_dates, forecast_values, color='rebeccapurple', marker='o', linestyle='--', linewidth=1, alpha=0.2)

# Format x-axis to show "Jan 2023", "Feb 2023", etc.
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

# Rotate the x-axis labels to make them more readable
plt.xticks(rotation=45)

# Add labels and title
#ax.set_title('Inflation Forecast vs Actual Inflation')
ax.set_ylabel('Monthly Inflation rates - Mean forecasts (%)')
ax.set_xlabel('')

# Show the plot
#plt.legend()
# Add a light gray grid
ax.grid(True, which='both', color='gainsboro', linestyle='-', linewidth=0.5)

# Only keep the x and y axes (bottom and left spines)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# You can also customize the remaining spines if needed (optional)
ax.spines['left'].set_linewidth(1)
ax.spines['bottom'].set_linewidth(1)

plt.tight_layout()
plt.show()
```

```python
sim_rem.columns = ["Aug 24", "Sep 24", "Oct 24", "Nov 24", "Dec 24", "Jan 25", "Feb 25"]
predictions_aug24.columns = ["Aug 24", "Sep 24", "Oct 24", "Nov 24", "Dec 24", "Jan 25", "Feb 25"]
```

```python
sim_rem.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
months = sim_rem.columns

# Create an empty list to store the data to plot
data_to_plot = []

# Loop through each month and append the data from both DataFrames
for month in months:
    data_to_plot.append(predictions_aug24[month].values)  # Data from df1
    data_to_plot.append(sim_rem[month].values)  # Data from df2

# Define positions for boxplots
positions = []
for i in range(len(months)):
    positions.append(i*2 + 0.9)  # Position for df1
    positions.append(i*2 + 1.1)  # Position for df2

# Create the boxplot
fig, ax = plt.subplots(figsize=(10, 6))


# Colors for df1 and df2
colors = ['darkcyan', 'palevioletred']
meanpointprops = [dict(marker=".",markersize=10,markeredgecolor="black", markerfacecolor="darkcyan"), dict(marker=".",markersize=10,markeredgecolor="black", markerfacecolor= "palevioletred")]

# Boxplot for df1 (left boxplots)
boxplot1 = ax.boxplot([data_to_plot[i] for i in range(0, len(data_to_plot), 2)],
                      positions=[positions[i] for i in range(0, len(positions), 2)],
                      widths=0.2, patch_artist=True, medianprops=dict(color="black"),
                      showmeans=True, meanprops=meanpointprops[0],
                      boxprops=dict(facecolor=colors[0], color='darkcyan'))

# Boxplot for df2 (right boxplots)
boxplot2 = ax.boxplot([data_to_plot[i] for i in range(1, len(data_to_plot), 2)],
                      positions=[positions[i] for i in range(1, len(positions), 2)],
                      widths=0.2, patch_artist=True, medianprops=dict(color="black"),
                      showmeans=True, meanprops=meanpointprops[1],
                      boxprops=dict(facecolor=colors[1], color='palevioletred'))

# Set x-ticks to be between the boxplots
ax.set_xticks([i * 2 + 1 for i in range(len(months))])
ax.set_xticklabels(months)


# Add a light gray grid
ax.grid(True, which='both', color='gainsboro', linestyle='-', linewidth=0.5)

# Only keep the x and y axes (bottom and left spines)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# You can also customize the remaining spines if needed (optional)
ax.spines['left'].set_linewidth(1)
ax.spines['bottom'].set_linewidth(1)

# Creating custom legend
source1_patch = mpatches.Patch(color='darkcyan', label='GPT-4o-mini ')
source2_patch = mpatches.Patch(color='palevioletred', label='REM')
plt.legend(handles=[source1_patch, source2_patch], loc='upper right')

# Adding labels and title
#plt.title("Expectativas de inflación mensual - IPC Nivel General")
plt.xlabel("")
plt.ylabel("Monthly prediction rates (%)")
plt.ylim(1, 6)  # Adjust based on your range
#plt.legend()
plt.tight_layout()
# Show the plot
plt.show()
```
